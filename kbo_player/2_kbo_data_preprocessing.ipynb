# merge하기

```python
import pandas as pd
```

```python
taza_tmp = pd.merge(taza,player_2022,left_on ='Key',right_on='Key',how = 'left').drop(['생년월일','입단년도','지명순위','연봉','신장/체중','입단 계약금','등번호'],axis = 1)
```

```python
player_2023.head(1)
```

```python
taza_sal = pd.merge(taza_tmp,player_2023,left_on ='Key',right_on='Key',how = 'left').drop(['생년월일','신장/체중','포지션_y','선수명_y','등번호','경력_y','지명순위','입단년도','Team_y','Team_x','연봉','입단 계약금','포지션_x','선수명_x'],axis = 1)
taza_sal
```

```python
taza_sal.head(1)
```

```python
taza_sal.columns = ['이름', '팀명', '타율', '경기수', '득점', '안타', '이루타', '삼루타', '홈런', '타점', 'Key',
       '경력', '연봉(원화)_2022', '연봉(원화)_2023']
```

```python
taza_sal['연봉차'] = taza_sal['연봉(원화)_2023']-taza_sal['연봉(원화)_2022']
taza_sal
```

```python
tusu_tmp = pd.merge(tusu,player_2022,left_on ='Key',right_on='Key',how = 'left').drop(['생년월일','신장/체중','입단 계약금','지명순위','입단년도','Team','연봉','선수명_y','등번호','포지션'],axis=1)
```

```python
tusu_sal = pd.merge(tusu_tmp,player_2023,left_on ='Key',right_on='Key',how = 'left').drop(['순위','생년월일','포지션','신장/체중','등번호','Team','입단년도','지명순위','경력_x','선수명','입단 계약금','연봉'],axis = 1)
```

```python
tusu_sal.columns = ['선수명', '팀명', '평균자책점', '경기', '승', '패', '세이브', '홀드', '승률', '이닝', '피안타',
       '홈런', '볼넷', '사구', '삼진', '실점', '자책점', '이닝당_출루허용률', 'Key', '연봉(원화)_2022',
       '경력', '연봉(원화)_2023']
```

```python
tusu_sal['연봉차'] = tusu_sal['연봉(원화)_2023'] - tusu_sal['연봉(원화)_2022']
```
