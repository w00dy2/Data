### DB접근해서 2023년 kbo연봉 데이터 가져오기

```python
import pymysql
try:                                                  #try구문을 사용하여 에러가 날 경우를 대비
    con = pymysql.connect(host='{DB url}', 
                          user='{username}', password='{userpassword}', 
                          charset='utf8', db='{dbname}')   #con은 DB와 연결을 해주는 역할
    cur = con.cursor()                                          #cur는 DB 쿼리를 실행시켜주는 역할
except Exception as e:
    print ("error ->", e)
```

```python
kbo_2023 = pd.read_sql_query("SELECT * FROM kbo_2023",con)
kbo_2023
```
