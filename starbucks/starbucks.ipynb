```python
import pandas as pd
```

```python
import pickle

with open("./starbucks.pkl", "rb") as f:
    star = pickle.load(f)
```

```python
star = pd.DataFrame(star)[["s_name","tel","fax" ,'open_dt',"sido_code","sido_name","gugun_code","gugun_name","addr","defaultimage","doro_address","lat","lot",]]
```

```python
import folium 
```

```python
import json
geo_data=json.load(open("./skorea_municipalities_geo_simple.json","r",encoding='utf-8'))
```

```python
star_map = folium.Map(location=[37.55,126.98], zoom_start=12)
for idx, row in star[['s_name', 'lat', 'lot']].iterrows(): #판다스의 로우값을 하나씩 가져온다.
    try:
        folium.Marker([float(row.lat), float(row.lot)], popup=row.s_name).add_to(star_map)
    except Exception as e: #예외처리
        print(e)
```

```python
import matplotlib.pyplot as plt
import numpy as np #숫자를 다루는 작업이라 넘파이, 판다스의 시각화기능이 맽플롯 변형이다
```

```python
data = star.groupby([''])['shopName'].count()

plt.style.use("ggplot")
plt.title('월별 오픈점포수')
x = np.arange(data.values.size)
name = data.index
values = data.values

plt.bar(x, values)
plt.xticks(x, name)

plt.show()
```
