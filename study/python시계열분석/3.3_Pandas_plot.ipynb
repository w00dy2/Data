색 선스타일 너비 등

```python
import pandas as pd
```

```python
df2 = pd.read_csv('df2.csv')
```

```python
df2.plot.line()
# 디폴트 값은 df2.plot()이다.
```

```python
df2.plot.line(figsize=(20,3))
```

```python
df2['c'].plot.line(figsize=(20,3),ls ='-')
```

```python
df2['c'].plot.line(figsize=(20,3),ls =':') #라인설정 점선
```

```python
df2['c'].plot.line(figsize=(20,3),ls =':',c= 'Red', lw = '5') #라인설정 점선
```

```python
title = 'My PLOT TITLE'
xlabel = 'My X Data'
ylabel = 'My Y Data'
```

```python
ax = df2['c'].plot.line(figsize=(20,3),ls =':',c= 'Red', lw = '5', title=title) #라인설정 점선
ax.set(xlabel = xlabel, ylabel = ylabel)
```

# 범례추가

```python
ax = df2.plot()
ax.legend(loc=0)
```

```python
ax = df2.plot()
ax.legend(loc = 0, bbox_to_anchor=(1.0,1.0)) # 범례를 외부에 놓기
```

```python
ax = df2.plot()
ax.legend(loc = 0, bbox_to_anchor=(0.5,0.5)) # 범례를 위치를 조절하여 사용이 가능하다.
```
