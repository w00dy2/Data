```python
import numpy as np
df = pd.DataFrame(np.random.randn(1000,2),columns=['a','b']) # a와b에 대한 임의의 값
```

```python
df.plot.scatter(x='a',y='b') # 산점도 플롯 생성하기
```

```python
df.plot.hexbin(x='a',y='b') # 육각형 모양의 빈
```

```python
df.plot.hexbin(x='a',y='b',gridsize=25) #사이즈 조절
```

```python
df.plot.hexbin(x='a',y='b',gridsize=25, cmap='Oranges') #색상변경
```
