streamlit은 데이터와 워크 그리고 프로젝트를 보여주기 위한 best framework.

수강 강의는 combination of import and theory together with hands on

즉, 이해를 돕기 위하여 실습도 진행한다.

### streamlit의 특징

1. great for career boost
2. easily - to - maintain
파이썬 빌트인

1. streamlit을 pip로 설치.
2. fast_kickoff에 포함된 app.py를 실행
    (python3 -m streamlit run ./fast_kickoff/app.py)



그렇게 하면 웹으로 접속이 가능하다.

streamlit의 포트번호는 8501이다. 하지만 몇몇 모듈이 설치되지 않았기 때문에 requirements.txt를 설치해준다.
